```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the CSV using second row as header
df = pd.read_csv("Season1_Shark_Tank_Data.csv", header=1)
df.columns = df.columns.str.strip()

# Define shark columns
shark_cols = ["Namita", "Anupam", "Vineeta", "Aman", "Peyush", "Ghazal"]

# Convert 'X' and blanks into 1 and 0 respectively
for shark in shark_cols:
    df[shark] = df[shark].apply(lambda x: 1 if str(x).strip().upper() == "X" else 0)

# Convert funding column from strings to numeric (lakhs)
def extract_funding_amount(funding_str):
    if isinstance(funding_str, str):
        match = re.search(r'Rs\.\s*([\d.]+)\s*lakhs', funding_str)
        if match:
            return float(match.group(1))
    return 0.0

df["Funding Amount (Lakhs)"] = df["Funding"].apply(extract_funding_amount)

# Total investment per shark
shark_totals = {}
for shark in shark_cols:
    shark_totals[shark] = (df[shark] * df["Funding Amount (Lakhs)"]).sum()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=list(shark_totals.keys()), y=list(shark_totals.values()), palette='coolwarm')
plt.title("Total Investments by Each Shark (in Lakhs)")
plt.xlabel("Sharks")
plt.ylabel("Total Funding (Lakhs)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

sector_funding = df.groupby("Industry/Genre")["Funding Amount (Lakhs)"].sum().sort_values(ascending=False)
top_sectors = sector_funding.head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_sectors, labels=top_sectors.index, autopct='%1.1f%%', startangle=140)
plt.title("🍕 Top 10 Industries by Total Funding Received")
plt.axis('equal')
plt.show()
```

```python
# Convert "Yes"/"No" in 'Deal' column to binary 1/0
df["Deal"] = df["Deal"].str.strip().str.lower().map({"yes": 1, "no": 0})

# Drop rows where Industry/Genre is missing (if any)
df = df[df["Industry/Genre"].notnull()]

# Get Top 10 Industries by Deal Count
top_deal_sectors = df[df["Deal"] == 1]["Industry/Genre"].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=top_deal_sectors.values, y=top_deal_sectors.index, palette='mako')
plt.title("📊 Top 10 Industries by Number of Deals")
plt.xlabel("Number of Deals")
plt.ylabel("Industry/Genre")
plt.show()
```

```python
import pandas as pd
import re

# Load the dataset (headers are on row 2)
df = pd.read_csv("Season1_Shark_Tank_Data.csv", header=1)
df.columns = df.columns.str.strip()

# Convert 'Deal' to binary
df["Deal"] = df["Deal"].astype(str).str.strip().str.lower().map({"yes": 1, "no": 0})

# Extract funding amounts
def extract_funding_amount(funding_str):
    if pd.isna(funding_str):
        return 0.0
    funding_str = str(funding_str).replace(",", "")
    match = re.search(r'(\d+\.?\d*)', funding_str)
    if match:
        return float(match.group(1))
    return 0.0

df["Funding_Amount_Lakhs"] = df["Funding"].apply(extract_funding_amount)

# Shark columns
shark_columns = ["Namita", "Anupam", "Vineeta", "Aman", "Peyush", "Ghazal"]
for shark in shark_columns:
    df[shark] = df[shark].astype(str).str.strip().str.upper().apply(lambda x: 1 if x == "X" else 0)

# Count sharks per deal
df["Num_Sharks"] = df[shark_columns].sum(axis=1).replace(0, 1)

# Pick a shark to see their investments
shark_name = "Aman"  # Change to "Vineeta", "Namita", etc. as needed

# Estimate the investment
df[f"{shark_name}_Investment_Lakhs"] = df.apply(
    lambda row: row["Funding_Amount_Lakhs"] / row["Num_Sharks"] if row[shark_name] == 1 else 0.0,
    axis=1
)
# Final table: Brand, Industry, Total Deal, Shark's Share
shark_investments = df[df[shark_name] == 1][
    ["Brand", "Industry/Genre", "Funding_Amount_Lakhs", f"{shark_name}_Investment_Lakhs"]
].reset_index(drop=True)

# Format currency columns to make it presentable
shark_investments["Funding_Amount_Lakhs"] = shark_investments["Funding_Amount_Lakhs"].apply(
    lambda x: f"₹{x:,.1f} Lakhs"
)
shark_investments[f"{shark_name}_Investment_Lakhs"] = shark_investments[f"{shark_name}_Investment_Lakhs"].apply(
    lambda x: f"₹{x:,.1f} Lakhs"
)

# Visualize
print(f"📊 {shark_name}'s Investment Breakdown:\n")
print(shark_investments)

shark_investments.to_csv(f"{shark_name}_Investments.csv", index=False)
```
